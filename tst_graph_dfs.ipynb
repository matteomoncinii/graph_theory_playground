```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an empty graph
G = nx.Graph()

# Define the edges based on the adjacency list
edges = [(0, 1), (0, 2), (0, 3),
         (1, 4), (1, 5),
         (2, 6), (2, 7),
         (3, 8),
         (4, 9),
         (5, 10),
         (6, 11),
         (7, 12),
         (7, 16),
         (6, 15),
         (8, 13),
         (9, 14),
         (11, 19),
         (12, 18),
         (13, 20),
         (14, 10),
         (16, 17),
         (17, 18),
         ]

# Add edges to the graph
G.add_edges_from(edges)

```

```python
# Draw the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=800, font_weight='bold', edge_color='gray')
plt.title("Graph with 15 Vertices")
plt.show()
```

## Pre-order Recursive DFS

```python
n = G.number_of_nodes()
marked = [False] * n
pre_order = []   

def visit(v, pre_order):
    #print(f"Visited: {v}")
    pre_order.append(v)

def dfs_recursive_pre(G, v):
    visit(v, pre_order)
    marked[v] = True
    for w in G.neighbors(v):
        if not marked[w]:  
            dfs_recursive_pre(G, w)  


dfs_recursive_pre(G, 0)
print(pre_order)
```

## Post-order Recursive DFS

```python
n = G.number_of_nodes()
marked = [False] * n
post_order = []   

def visit(v, post_order):
    #print(f"Visited: {v}")
    post_order.append(v)

def dfs_recursive_pre(G, v):
    marked[v] = True
    for w in G.neighbors(v):
        if not marked[w]:  
            dfs_recursive_pre(G, w)  
    visit(v, post_order)


dfs_recursive_pre(G, 0)
print(post_order)
```
