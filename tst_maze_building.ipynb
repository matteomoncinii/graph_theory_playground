```python
import networkx as nx
import matplotlib.pyplot as plt
from pprint import pprint
from random import randint
```

```python
rows, cols = 8, 8

# Create an empty graph
G = nx.Graph()

# Add nodes
for i in range(rows):
    for j in range(cols):
        G.add_node((i, j), pos=(j, -i)) 

# Add edges
for i in range(rows):
    for j in range(cols):
        if i > 0:
            G.add_edge((i, j), (i-1, j))  # add up
        if i < rows - 1:
            G.add_edge((i, j), (i+1, j))  # add down
        if j > 0:
            G.add_edge((i, j), (i, j-1))  # add left
        if j < cols - 1:
            G.add_edge((i, j), (i, j+1))  # add right


# Define entry and exit points
entry_node = (0, 0)
exit_node = (rows-1, cols-1)

# Custom node colors
node_colors = []
for node in G.nodes():
    if node == entry_node:
        node_colors.append("green")  
    elif node == exit_node:
        node_colors.append("red")    
    else:
        node_colors.append("lightblue")

custom_labels = {entry_node: "ENTER", exit_node: "EXIT"}
```

```python
# Randomly remove some edges

removed = []

for n in range(int(len(list(G.edges))*0.25)):
    i = randint(0, len(list(G.edges)))
    node1, node2 = list(G.edges)[i]
    G.remove_edge(u=node1, v=node2)
```

```python
# Draw the maze
plt.figure(figsize=(5, 5))
pos = nx.get_node_attributes(G, 'pos')
nx.draw(G, pos, with_labels=True, labels=custom_labels,
        node_size=400, node_color=node_colors, edge_color='gray',
        font_weight='bold', font_color='black')

plt.gca().set_aspect('equal')
plt.axis('off')
plt.show()
```
